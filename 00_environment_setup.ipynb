```python
from scipy import stats

# PDF & CDF of normal
dist = stats.norm(loc=0, scale=1)
dist.pdf(0), dist.cdf(1)
```

```python
# Generate x-values
x = np.linspace(-3, 3, 100)

# Compute PDF and CDF
pdf_values = dist.pdf(x)
cdf_values = dist.cdf(x)

# Plot PDF
plt.figure(figsize=(8, 4))
plt.plot(x, pdf_values, label="PDF", color="blue")
plt.title("Probability Density Function (PDF)")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.grid(True)

# Plot CDF
plt.figure(figsize=(8, 4))
plt.plot(x, cdf_values, label="CDF", color="red")
plt.title("Cumulative Distribution Function (CDF)")
plt.xlabel("x")
plt.ylabel("Cumulative Probability")
plt.legend()
plt.grid(True)

plt.show()
```

```python
samples = dist.rvs(size=1000)
pd.Series(samples).plot(kind="hist", bins=40, density=True)
```

```python
# One-sample t-test: is mean=0?
stats.ttest_1samp(samples, popmean=0)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a matrix B
B = np.array([[3, 2, 2], [2, 3, -2]])

# Compute SVD
U, S, Vt = np.linalg.svd(B)

# Convert S to a diagonal matrix
S_matrix = np.diag(S)

# Create subplots for visualization
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

# Plot the original matrix B
axes[0].imshow(B, cmap="viridis", aspect="auto")
axes[0].set_title("Original Matrix B")

# Plot singular values S (as a diagonal matrix)
axes[1].imshow(S_matrix, cmap="viridis", aspect="auto")
axes[1].set_title("Singular Values (S)")

# Plot V (Right Singular Vectors)
axes[2].imshow(Vt, cmap="viridis", aspect="auto")
axes[2].set_title("V (Right Singular Vectors)")

# Adjust layout and display plot
plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2).fit(B)
pca.explained_variance_ratio_
```

```python
import numpy as np
from scipy import stats

# Generate sample data (normally distributed around 5)
np.random.seed(42)
samples = np.random.normal(loc=5, scale=2, size=100)

# Perform one-sample t-test (testing if mean = 5)
t_statistic, p_value = stats.ttest_1samp(samples, popmean=5)

# Print results
print(f"T-Statistic: {t_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation:
if p_value < 0.05:
    print("Reject the null hypothesis: The sample mean is significantly different from 5.")
else:
    print("Fail to reject the null hypothesis: The sample mean is likely close to 5.")
```

```python
import numpy as np

# Assume 'samples' is the dataset from the t-test
sample_mean = np.mean(samples)

print(f"Sample Mean: {sample_mean:.4f}")
```
